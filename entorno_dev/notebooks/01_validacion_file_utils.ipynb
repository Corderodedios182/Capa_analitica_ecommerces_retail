### Pruebas para STAGING : Limpieza estructural.

Creamos una capa técnica previa a **STAGING** usando Python para **leer, homologar y unificar múltiples archivos RAW por dominio**. Consolidamos **clientes (3 fuentes)**, **órdenes (2 fuentes)** y **catálogo (1 fuente)** en **DataFrames canónicos únicos**, listos para persistirse en **SQLite** y ser consumidos por **dbt** como fuentes STAGING limpias, trazables y reutilizables.


```python
from scripts.ingestion.unify_clientes import unify_clientes
from scripts.ingestion.unify_ordenes import unify_ordenes

df_clientes_cdmx = read_source("clientes_cdmx")
df_clientes_gdl = read_source("clientes_gdl_mty")
df_clientes_resto = read_source("clientes_resto")

df_clientes = unify_clientes(df_clientes_cdmx, df_clientes_gdl, df_clientes_resto)

#Validacion de registros hace match con el número de registros desde raw.
df_clientes.groupby("source_name").count()
```

```python
df_orders_hist = read_source("ordenes_2022_2023")
df_orders_2024 = read_source("ordenes_2024")

df_ordenes = unify_ordenes(df_orders_hist, df_orders_2024)

df_ordenes.groupby("source_name").count()
```
